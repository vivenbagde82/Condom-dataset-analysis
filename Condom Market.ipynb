```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Year', y='Market Size (USD Million)', marker='o')
plt.title('Market Size Over the Years')
plt.xlabel('Year')
plt.ylabel('Market Size (USD Million)')
plt.show()
```

```python
brand_counts = df['Brand Name'].value_counts().reset_index()
brand_counts.columns = ['Brand Name', 'Count']

material_counts = df.groupby(['Brand Name', 'Material Type']).size().unstack(fill_value=0).reset_index()

brand_counts = brand_counts.merge(material_counts, on='Brand Name')

melted_df = brand_counts.melt(id_vars=['Brand Name', 'Count'], value_vars=['Latex', 'Non-latex'], var_name='Material Type', value_name='Material Count')

plt.figure(figsize=(12, 8))
ax = sns.barplot(data=melted_df, x='Brand Name', y='Count', color='skyblue')  # Barplot for total count of each brand
plt.title('Distribution of Brand Names and Material Types')
plt.xlabel('Brand Name')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')

ax2 = ax.twinx()  # Create a second y-axis
sns.lineplot(data=melted_df[melted_df['Material Type'] == 'Latex'], x='Brand Name', y='Material Count', marker='o', color='red', label='Latex', ax=ax2)
sns.lineplot(data=melted_df[melted_df['Material Type'] == 'Non-latex'], x='Brand Name', y='Material Count', marker='o', color='green', label='Non-Latex', ax=ax2)

ax.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Company Involved')
plt.title('Count of Companies Involved')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='Company Involved', hue='Event Details')
plt.title('Event Details by Company Involved')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Event Details')
plt.show()
```

```python
region_brand_counts = df.groupby(['Region', 'Brand Name']).size().unstack(fill_value=0)
plt.figure(figsize=(12, 8))
sns.heatmap(region_brand_counts, annot=True, cmap='YlGnBu')
plt.title('Brand Popularity/Market Share')
plt.xlabel('Brand Name')
plt.ylabel('Regions')
plt.tight_layout()
plt.show()
```
